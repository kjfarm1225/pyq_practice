```python
import pandas as pd
df = pd.read_csv('service_sample.csv')
df
```

```python
print(df)
```

```python
df["個数x期間"] = df["個数"] * df["期間"]
display(df)
```

```python
price_map = {
    "通常": 6000,
    "学割": 3000
}
df["単価"] = df["プラン"].map(price_map)
```

```python
df
```

```python
df["契約金額"] = df["個数x期間"] * df["単価"]
df
```

```python
def size_category(x):
    if x < 100000:
        return "小口"
    elif x < 200000:
        return "重要"
    elif x < 600000:
        return "中口"
    else:
        return "大口"
df["契約規模"] = df["契約金額"].apply(size_category)
display(df)
```

```python
df_normal = df[df["プラン"] == "通常"]
```
