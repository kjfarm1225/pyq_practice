```python
import pandas as pd

df = pd.read_csv(
    "visitors_sample.csv",
    index_col="日付",
    parse_dates=["日付"]
)
df.head()
```

```python
monthly = df.resample("ME").sum()
monthly
```

```python
monthly_mean = df.resample("ME").mean()
monthly_mean
```

```python
df.resample("ME")["来店者数"].max()
```

```python
df.resample("ME")["売上"].sum()
```

```python
df.resample("ME").agg(["sum", "mean", "max"])
```

```python
monthly = df.resample("ME").sum()
```

```python
monthly["平均売上"] = monthly["売上"] / monthly["来店者数"]
monthly
```

```python
monthly = df.resample("ME").sum()
monthly["不良率"] = monthly["来店者数"] / monthly["売上"]
monthly
```
