```python
import pandas as pd

df = pd.read_csv("service_detail_step4_sample.csv")

summary_df = pd.DataFrame(index=df["顧客名"].unique())
grouped = df.groupby("顧客名")

def count_by_category(x, category):
    return (x == category).sum()

for category in ["小口", "中口", "大口"]:
    summary_df[f"契約回数_{category}"] = grouped["契約規模"].agg(
        count_by_category, category=category
    )
summary_df
```

```python
import pandas as pd
df = pd.read_csv("service_detail_step4_sample.csv")
grouped = df.groupby("顧客名")

summary_df = pd.DataFrame()
summary_df["契約回数"] = grouped.size()
summary_df["契約金額合計"] = grouped["契約金額"].sum()
summary_df["初回レンタル開始日"] = grouped["レンタル開始日"].min()

def count_by_category(x, category):
    return (x == category).sum()

for category in ["小口", "中口", "大口"]:
    summary_df[f"契約回数_{category}"] = grouped["契約規模"].agg(
        count_by_category, category=category
    )

summary_df
```

```python
summary_df = summary_df.sort_values(
    ["契約金額合計", "初回レンタル開始日"],
    ascending=False
)
summary_df
```
