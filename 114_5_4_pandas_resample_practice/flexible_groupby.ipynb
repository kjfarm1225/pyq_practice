```python
import pandas as pd

df = pd.read_csv("earnings_sample.csv", index_col="日付", parse_dates=["日付"])
df
```

```python
df.resample("W")["売上（万円）"].sum()
```

```python
df["売上（万円)_ビン分割"] = pd.cut(
    df["売上（万円）"],
    [0, 10, 20, 30, 40, 50],
    right=False,
)
df
```

```python
df.groupby("支店名")["売上（万円）"].sum()
```

```python
df.groupby(["支店名", "売上（万円)_ビン分割"], observed=False).size()
```

```python
df.groupby(["支店名", "売上（万円)_ビン分割"], observed=False)["売上（万円）"].mean()
```

```python
df[df["支店名"]== "B支店"].groupby("売上（万円)_ビン分割", observed=False).size()
```
