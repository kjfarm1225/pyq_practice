```python
import pandas as pd
df = pd.read_csv("service_detail_after_sample.csv")
df.head()
```

```python
grouped = df.groupby("顧客名")
grouped
```

```python
df.groupby("顧客名").size()
```

```python
summary_df = grouped["個数x期間"].agg(["size", "sum"])
summary_df
```

```python
summary_df = summary_df.rename(
    columns={
        "size": "契約回数",
        "sum": "個数x期間の合計"
    }
)
summary_df
```

```python
summary_df["契約金額の合計"] = grouped["契約金額"].sum()
summary_df
```

```python
bins = [-float("inf"), 100000, 600000, 1200000, 2400000, float("inf")]
summary_df["契約金額区分"] = pd.cut(
    summary_df["契約金額の合計"],
    bins
)
summary_df
```
